# 1.2 Data Visualisation in Python

```python
fig, ax = plt.subplots(figsize=(17,7))
data.groupby(['REGION','QUALITY_PRODUCT']).sum()['TOTAL_PROFIT'].unstack().plot(ax=ax, marker='o',markersize=10)
ax.set_ylabel('Amount of TOTAL PROFIT');
```

```python
data = pd.read_csv('100000.csv')
data['SHIP_DATE']
```

```python
fig, ax = plt.subplots(figsize=(17,7))
data.groupby(['SHIP_DATE','QUALITY_PRODUCT']).sum()['UNITS_SOLD'].unstack().plot(ax=ax, marker='o',markersize=10)
ax.set_ylabel('Amount of TOTAL PROFIT');
```

```python
# Quantity country in Region
ax=data.groupby(['REGION']).sum()['TOTAL_PROFIT'].plot.bar(title='Total Profit each Region')
t=len(data)
for p in ax.patches:
    ax.annotate('%{:1.1f}'.format(p.get_height()/t), (p.get_x()+0.1, p.get_height()+5),ha = 'center');
```

```python
#Quantity type of products
data.groupby(['REGION']).sum()['TOTAL_PROFIT'].plot.pie(autopct = '%1.1f%%',shadow=True,startangle=90)
# Unsquish the pie.
plt.axis('equal')
fig = plt.gcf()
fig.set_size_inches(8,8)
```

```python
# Statistics TOTAL PROFIT 
sns.distplot(data['TOTAL_PROFIT'],bins=10, hist_kws={"alpha": 1,}).set(xlabel='Total Profit', ylabel='Count');
```

```python
# Quantity LEVEL type ORDER
sns.set(style="darkgrid")
plt.figure(figsize=(7, 4))
ax =sns.countplot(data['QUALITY_PRODUCT'])
t=len(data)*1
plt.title("QUALITY_PRODUCT")
plt.xlabel('Count level Product')
for p in ax.patches:
    ax.annotate('%{:0.3f}'.format(100*p.get_height()/t), (p.get_x()+0.1, p.get_height()+5))
```

```python
#Quantity type of products
data['QUALITY_PRODUCT'].value_counts().plot.pie(autopct = '%.3f%%',shadow=True,startangle=90)
# Unsquish the pie.
plt.axis('equal')
fig = plt.gcf()
fig.set_size_inches(8,8)
```

```python
#Quantity type of products
data['ITEM_TYPE'].value_counts().plot.pie(autopct = '%.3f%%',shadow=True, startangle=140)
# Unsquish the pie.
plt.axis('equal')
fig = plt.gcf()
fig.set_size_inches(8,8)
```
